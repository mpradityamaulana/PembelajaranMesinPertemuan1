```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
myArray = np.array([1,2,3])
plt.plot(myArray)
plt.xlabel('Axis x')
plt.ylabel('Axis y')
plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])

plt.scatter(x,y)
plt.xlabel("Axis x")
plt.ylabel("Axis y")
plt.show()
```
